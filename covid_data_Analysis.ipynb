```python
import pandas as pd
import plotly.express as px
%matplotlib inline
```

# Data Set loading for visualization

```python
url = "https://covid.ourworldindata.org/data/owid-covid-data.csv"
data = pd.read_csv(url,sep=',')
data.head(5)
```

```python
 data.to_csv('covid_data.csv',';')
```

```python
data.columns
```

```python
b_Ger = data['location']=='Germany'
b_Ind = data['location']=='India'
b_Aus = data['location']=='Australia'
b_all = b_Ger|b_Ind|b_Aus
sum(b_all)
```

```python
data_final = data[b_all]
data_final['covid_rate'] = data_final['total_cases']/data_final['population']
```

```python
data_final['vacc_rate'] = data_final['people_fully_vaccinated']/data_final['population']
```

# Data Plotting
Delivery1 and Delivery2

# Plotting Vaccination rate overtime

```python
fig = px.line(data_final, x="date", y="vacc_rate", color='location',title='Vaccination rate')
fig.show()
```

# Plotting Relative covid cases overtime

```python
fig = px.line(data_final, x="date", y="covid_rate", color='location',title='Relative COVID Cases')
fig.show()
```
